```python
##缺失值填补

import pandas as pd
data=pd.read_csv(r"/Users/lijianwei/Desktop/菜菜sklearn/day08_Narrativedata.csv",index_col=0)
```

```python
data.head()
```

```python
data.info()
```

```python
#用0，平均值，中位数填补缺失值
from sklearn.impute import SimpleImputer
data["Age"].unique()
```

```python
Age=data["Age"].values.reshape(-1,1)
```

```python
import numpy as np
#用0填补缺失值
imp_0=SimpleImputer(missing_values=np.nan,strategy="constant",fill_value=0)
imp_0=imp_0.fit_transform(Age)

#用平均值填补缺失值
imp_mean=SimpleImputer(missing_values=np.nan,strategy="mean").fit_transform(Age)

#用中位数填补缺失值
imp_median=SimpleImputer(missing_values=np.nan,strategy="median").fit_transform(Age)
```

```python
imp_0
```

```python
imp_mean
```

```python
imp_median
```

```python
data["Age"]=imp_median
```

```python
data.info()
```

```python
#缺失值比较少值 我们可以直接删除
data=data.dropna()
```

```python
data.info()
```

```python
data
```

```python
#处理分类数据特征
#preprocessing.LabelEncoder: 标签专用，能够将标签转换为分类数值
from sklearn.preprocessing import LabelEncoder
y=data.iloc[:,-1]
le=LabelEncoder()
le.fit(y)
```

```python
le.classes_
```

```python
label=le.transform(y)
```

```python
label
```

```python
le.inverse_transform(label)
```

```python

#对于名义变量，我们需要使用独热编码（哑变量） preprocessing.OnHotEncoder
from sklearn.preprocessing import OneHotEncoder
X=data.iloc[:,1:-1]
enc=OneHotEncoder().fit(X)
result=enc.transform(X).toarray()
result
```

```python
#获取特征名称 get_feature_names
enc.get_feature_names()
```

```python
newdata=pd.concat([data,pd.DataFrame(result)],axis=1)
newdata
```

```python
newdata.drop(["Sex","Embarked"],axis=1,inplace=True)
newdata.columns=["Age","Survived",'Female', 'Male', 'Embarked_C', 'Embarked_Q', 'Embarked_S']
newdata.head()
```

```python
#处理连续性特征：二值化与分段 preprocessing.Binarizer
data_2=data.copy()
from sklearn.preprocessing import Binarizer
X=data_2.iloc[:,0].values.reshape(-1,1)
transformer=Binarizer(threshold=38).fit_transform(X)#######threshold 为阈值
transformer
```

```python
#处理连续性变量，能够将连续性变量排序后按顺序分箱处理  preprocessing.KBinsDiscretizer
from sklearn.preprocessing import KBinsDiscretizer

X=data.iloc[:,0].values.reshape(-1,1)
est=KBinsDiscretizer(n_bins=3,encode='ordinal',strategy='uniform')     ##encode编码方式：onehot、ordinal    strategy用来定义箱宽方式：uniform(等宽分箱)、quantile（等位分箱）、kmeans（聚类分箱）
est.fit_transform(X)
```

```python
set(est.fit_transform(X).ravel())  ##ravel（）降纬  
```
